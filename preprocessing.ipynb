```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('hotel_bookings.csv')
dataset.head()
```

```python
dataset.columns
```

```python
dataset.drop(['company', 'arrival_date_week_number', 'reservation_status',
              'reservation_status_date', 'arrival_date_month', 'arrival_date_year',
              'arrival_date_day_of_month'], axis=1, inplace=True)
```

```python
dataset.dropna(subset=['country'], inplace=True)
```

```python
dataset = dataset[dataset['adr'] != 5400]
```

```python
# for agent that appears less than 100 times, replace by a number that corresponds to others
agent_count = dataset['agent'].value_counts()
agent_count = agent_count[agent_count < 100]
agent_count = agent_count.index.tolist()

# search the highest number in agent
max_agent = dataset['agent'].max()
max_agent = max_agent + 1

dataset['agent'].replace(agent_count, max_agent, inplace=True)

 
```

```python
# count agent
dataset['agent'].value_counts()
```

```python
# create a column is reserved_room_type same than assigned_room_type
dataset['is_room_same'] = np.where(dataset['reserved_room_type'] == dataset['assigned_room_type'], 1, 0)
```

```python
dataset.head()
```

```python
# export dataset
dataset.to_csv('hotel_bookings_clean.csv', index=False)
```
